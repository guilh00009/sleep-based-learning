```python
# Import the sleep engine which contains all the core logic
import sleep_engine
import gradio as gr

# Initialize the system (loads the model, sets up logging)
sleep_engine.initialize_system()
```

```python
# Launch the Gradio Interface
print("Launching Gradio interface...")
iface = gr.ChatInterface(
    fn=sleep_engine.chat_with_gemma,
    title="Gemma-3 4B-IT - Dream Engine (Debug Mode)",
    description="Talk to the model or type `/sleep`. Logs are in 'app.log'",
    theme="soft"
)
iface.launch(server_name="0.0.0.0", server_port=7860, share=True)
```

```python
# Optional: Inspect Conversation History
history = sleep_engine.conversation_history
print(f"The conversation has a total of {len(history)} messages.")
print("-" * 50)

for i, message in enumerate(history):
    role = message['role'].capitalize()
    content = message['content']
    print(f"Turn {i+1} - [{role}]:\n{content}\n")
    print("-" * 20)
```
